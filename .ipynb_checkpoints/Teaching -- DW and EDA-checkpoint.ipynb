```python
me = pd.merge(df, rz, left_on='SISNumber', right_on='SIS Number').drop('SIS Number', axis=1)
me
```

```python
#I'm too excited to not do some preliminary bar graphs, just for giggles!

import matplotlib.pyplot as plt

race_counts = me.groupby('Race')['Race'].count().sort_values()
```

```python
print(race_counts)
```

```python
race_counts.index
```

```python
plt.style.use('ggplot')

races = race_counts.index
counts = race_counts

fig, ax = plt.subplots(figsize=(10,6))

ax.bar(races, counts)
ax.set_ylabel('Count')
ax.set_title('Total Number of Students Taught by Race')
plt.xticks(rotation=90)
plt.show()
```
